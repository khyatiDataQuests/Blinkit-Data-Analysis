# <p style="font-family:newtimeroman;font-size:130%;text-align:center;color:#06445e;">1. Data Loading & Cleaning</p>

```python
orders=pd.read_csv('blinkit_orders.csv')
order_items=pd.read_csv('blinkit_order_items.csv')
customers=pd.read_csv('blinkit_customers.csv')
customer_feedback=pd.read_csv('blinkit_customer_feedback.csv')
delivery=pd.read_csv('blinkit_delivery_performance.csv')
marketing=pd.read_csv('blinkit_marketing_performance.csv')
```

# <p style="font-family:newtimeroman;font-size:100%;text-align:left;color:#06445e;">Changing Datatypes</p>
#### ***Data types were corrected to ensure consistency and accurate analysis across datasets. Date fields were converted to datetime format for time-based insights, and numerical fields were validated for reliable calculations.***

```python
# Orders
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['promised_delivery_time'] = pd.to_datetime(orders['promised_delivery_time'])
orders['actual_delivery_time'] = pd.to_datetime(orders['actual_delivery_time'])

# Customers
customers['registration_date'] = pd.to_datetime(customers['registration_date'])

# Feedback
customer_feedback['feedback_date'] = pd.to_datetime(customer_feedback['feedback_date'])

# Delivery
delivery['promised_time'] = pd.to_datetime(delivery['promised_time'])
delivery['actual_time'] = pd.to_datetime(delivery['actual_time'])

# Marketing
marketing['date'] = pd.to_datetime(marketing['date'])
```

# <p style="font-family:newtimeroman;font-size:100%;text-align:left;color:#06445e;">Removing Duplicates</p>
#### **_Removing duplicates is an important step in data wrangling. Duplicates can lead to incorrect insights by skewing statistics like mean, median, and standard deviation, and by over-representing certain data points. Removing them ensures accurate and reliable analysis._**

```python
for name, df in datasets.items():
    print(f"\n{name.upper()} DUPLICATES")
    
    dup_count = df.duplicated().sum()
    print("Total duplicate rows:", dup_count)
```

## ***1. Order & Revenue***
### ***What is the distribution of order values?***


```python


plt.figure(figsize=(8,5))
sns.histplot(orders['order_total'],kde=True, bins=30)

plt.title("Distribution of Order Values")
plt.xlabel("Order Total")
plt.ylabel("Frequency")

plt.show()
```

### ***Answer 1:*** 
-  ***The distribution of order values is `positively skewed`, with a long tail of high-value transactions. This indicates that while the majority of orders are of moderate value, a small segment of high-value orders contributes disproportionately to overall revenue.***

- ***This pattern suggests the presence of premium purchasing behavior, highlighting an opportunity for targeted strategies such as personalized offers, upselling, or loyalty programs to maximize revenue from high-value customers.***

## ***Time Series Analysis***

### ***1. How does order volume change over time?***

```python

orders['order_date_only'] = orders['order_date'].dt.date

# Orders per day
daily_orders = orders.groupby('order_date_only').size()

plt.figure(figsize=(10,5))
daily_orders.plot()

plt.title("Order Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.show()
```

### ***Answer 2:*** 
- ***Order volume over time shows noticeable fluctuations, indicating variability in daily demand patterns. Certain periods exhibit spikes in order activity, suggesting peak demand days, while other periods reflect relatively lower engagement.***

- ***This variability highlights the importance of demand forecasting and resource planning. Aligning inventory and delivery capacity with peak demand periods can significantly improve operational efficiency and customer satisfaction..***

### ***2. What is the weekly order count?***

```python
# Create week number
orders['week'] = orders['order_date'].dt.isocalendar().week

weekly_orders = orders.groupby('week').size()

plt.figure(figsize=(10,5))
plt.plot(weekly_orders.index, weekly_orders.values, marker='o')

plt.title("Weekly Order Volume")
plt.xlabel("Week Number")
plt.ylabel("Number of Orders")

plt.show()
```

### ***Answer 2:***
- ***Weekly order volume shows clear fluctuations across different weeks, indicating varying customer demand.***

- ***There is a noticeable increase in orders during the mid-weeks, followed by a decline towards the later weeks, suggesting changing demand patterns over time.***

### ***3. Are there any seasonal patterns in orders or revenue?***

```python
# Create month column
orders['month'] = orders['order_date'].dt.month_name()

# Correct month order
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

# Orders per month
monthly_orders = orders.groupby('month').size().reindex(month_order)

# Revenue per month
monthly_revenue = orders.groupby('month')['order_total'].sum().reindex(month_order)

import matplotlib.pyplot as plt

# orders plot
plt.figure(figsize=(10,4))
plt.plot(monthly_orders.index, monthly_orders.values, marker='o')
plt.title("Monthly Orders")
plt.xticks(rotation=45)
plt.show()

# revenue plot
plt.figure(figsize=(10,4))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o')
plt.title("Monthly Revenue")
plt.xticks(rotation=45)
plt.show()
```

### ***Answer 3:***
- ***The monthly revenue trend shows a `rise from early months to mid-year`, followed by a `decline towards the end of the year`. However, this pattern does not strongly align with typical seasonal expectations where demand often increases in the later months.***

- ***This suggests that the dataset may not exhibit strong real-world seasonality or that demand is influenced by factors not captured in the data. The observed variation appears moderate rather than strongly seasonal.***

- ***Therefore, while there are fluctuations in revenue, there is no clear and consistent seasonal pattern. Additional data or longer time periods would be required to confirm true seasonality.***

## ***Delivery Performance and Analysis***

### ***1. What is the On-Time vs Late Delivery Rate?***

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(8,5))
sns.countplot(data=orders, x='delivery_status')

plt.title("Delivery Status Distribution")
plt.xlabel("Delivery Status")
plt.ylabel("Count")

plt.show()
```

### ***Answer 1:***
- ***Most orders are `delivered on time`, indicating `strong overall operational efficiency`. A smaller portion of orders are slightly or significantly delayed, suggesting some logistical bottlenecks.***
- ***These delays could impact customer satisfaction and retention. Monitoring delivery performance and optimizing delivery partner operations can help reduce delays and improve service quality***

### ***2. Does delivery delay affect customer ratings?*** 

```python

# Merging datasets
merged = pd.merge(orders, customer_feedback, on='order_id', how='inner')

 
merged['delay'] = (
    merged['actual_delivery_time'] - merged['promised_delivery_time']
).dt.total_seconds() / 60


plt.figure(figsize=(8,5))
sns.boxplot(x=merged['rating'], y=merged['delay'])

plt.title("Delivery Delay vs Customer Rating")
plt.xlabel("Rating")
plt.ylabel("Delay (minutes)")
plt.show()
```

### ***Answer 2:***
- ***A clear relationship is observed between delivery delay and customer ratings, where `higher delays tend to be associated with lower ratings`.***

- ***This indicates that delivery timeliness `plays a critical role in shaping customer satisfaction`. Improving delivery efficiency can directly contribute to better customer experience and higher ratings.***

## ***Customer Behavior Analysis***

### ***1. What is the distribution of customer segments?***

```python
customers['customer_segment'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
```

### ***Answer 1:***
- ***The distribution of customer segments shows how customers are categorized based on their engagement level.***

- ***A higher proportion of premium or regular customers indicates strong customer engagement, while a  inactive segment may highlight potential churn risk. This segmentation helps in designing targeted marketing and retention strategies.***

### ***2. Do different customer segments show differences in order value?***

```python
merged = orders.merge(customers, on='customer_id')

plt.figure(figsize=(8,5))
sns.boxplot(x=merged['customer_segment'], y=merged['order_total'])

plt.title("Order Value by Customer Segment")
plt.show()
```

- ***The distribution of order values across different customer segments appears `relatively similar`, with no segment showing a significantly higher median order value.***

- ***While there are slight variations, the overlap between segments suggests that customer spending behavior is `fairly consistent` across all groups.***

- ***This indicates that revenue is not heavily dependent on a single customer segment, and the business can focus on improving engagement and retention across all segments rather than targeting only one group.***

### ***3. What is the repeat customer rate?***

```python
## total customers
total_customers = customers['customer_id'].nunique()

## repeat customers 
repeat_customers = customers[customers['total_orders'] > 1]['customer_id'].nunique()

## repeat rate
repeat_rate = (repeat_customers / total_customers) * 100

print("Repeat Customer Rate:", repeat_rate)
```

### ***Answer 3:*** 
- ***A significant portion of customers that is `94.199` are repeat buyers, indicating strong customer retention and satisfaction.***

- ***High repeat rates suggest that customers find value in the service, while lower repeat rates may indicate the need for improved customer engagement strategies such as loyalty programs or personalized offers.***

## ***Product and Basket Analysis***

### ***1. Which products generate the highest revenue?***

```python
## creating revenue column
order_items['revenue'] = order_items['quantity'] * order_items['unit_price']

## top products by revenue
top_products = order_items.groupby('product_id')['revenue'].sum().sort_values(ascending=False).head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top_products.plot(kind='bar')

plt.title("Top 10 Products by Revenue")
plt.xlabel("Product ID")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

### ***Answer 1:***
- ***`Product 57036` is the clear revenue leader, indicating strong market demand, pricing effectiveness, or both—making it a `key asset` for revenue sustainability.***
- ***`Products 34186` and `880510` form the second tier of high performers, contributing significantly and demonstrating a stable revenue base beyond the top product.***
- ***Revenue contribution shows a gradual decline across the top 10 products, suggesting a well-distributed portfolio rather than heavy reliance on a single product.***
- ***Strategic focus should be twofold: `protect and scale top-performing products` while identifying optimization opportunities (pricing, promotion, bundling) for mid-tier products to close the revenue gap.***

### ***2. What is the distribution of quantity per order?***

```python


plt.figure(figsize=(8,5))
sns.histplot(order_items['quantity'], bins=20)

plt.title("Distribution of Quantity per Order")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

### ***Answer 2:***
- ***Order quantities are highly concentrated `between 1 and 3 units`, indicating that most customers purchase in small volumes per transaction.***

- ***From a business perspective, this indicates an opportunity to increase average order value (AOV) through some strategies.***

### ***3. Are few products dominating total sales?***

```python
# Percentage contribution
product_revenue = order_items.groupby('product_id')['revenue'].sum()

top_10 = product_revenue.sort_values(ascending=False).head(10)
top_10_share = (top_10.sum() / product_revenue.sum()) * 100

print("Top 10 Products Contribution (%):", top_10_share)
```

### ***Answer 3:***
- ***The top 10 products contribute approximately `10% of total revenue`, indicating that revenue is not heavily concentrated among a few items but is instead distributed across a wide range of products.***

- ***From a business perspective, this is a `positive sign` as it lowers risk; however, identifying and promoting higher-performing products could further optimize revenue growth.***

## ***Marketing Performance Analysis***

### ***1. Which campaigns generate the highest revenue?***

```python
# Top campaigns by revenue
top_campaigns = marketing.groupby('campaign_name')['revenue_generated'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_campaigns.plot(kind='bar')

plt.title("Top Campaigns by Revenue")
plt.xlabel("Campaign")
plt.ylabel("Revenue Generated")
plt.xticks(rotation=45)
plt.show()
```

### ***Answer 1:***
- ***`Referral Program` is the top-performing campaign, generating the highest revenue, indicating strong effectiveness of word-of-mouth and customer-driven acquisition strategies.***
- ***`New User Discount` and `Email Campaign` closely follow, highlighting that both customer acquisition incentives and direct marketing channels are key revenue drivers.***
- ***Revenue across campaigns is `relatively uniform with marginal differences`, suggesting a well-balanced marketing mix where multiple campaigns contribute consistently.***


### ***2. How efficient are campaigns (Conversion Rate & ROAS)?***

```python
# Create metrics
marketing['conversion_rate'] = marketing['conversions'] / marketing['clicks']

# Average metrics by campaign
campaign_metrics = marketing.groupby('campaign_name')[['conversion_rate','roas']].mean()

# Top campaigns by ROAS
top_roas = campaign_metrics.sort_values(by='roas', ascending=False).head(10)

plt.figure(figsize=(10,5))
top_roas['roas'].plot(kind='bar')

plt.title("Top Campaigns by ROAS")
plt.xlabel("Campaign")
plt.ylabel("ROAS")
plt.xticks(rotation=45)
plt.show()
```

### ***Answer 3:***
- ***The ROAS across different marketing campaigns appears `relatively consistent`, with all campaigns delivering similar returns on ad spend.***

- ***While this reflects stability in campaign performance, it also highlights an opportunity to further optimize and differentiate campaigns to achieve higher returns. Testing new strategies, targeting methods, or creatives could help improve overall marketing effectiveness.***

### ***4. What is the customer growth trend over time?***

```python

customers['registration_date'] = pd.to_datetime(customers['registration_date'])
customers = customers.sort_values('registration_date')

customers['cumulative_customers'] = range(1, len(customers) + 1)


import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(customers['registration_date'], customers['cumulative_customers'])

plt.title("Customer Growth Over Time")
plt.xlabel("Date")
plt.ylabel("Total Customers")
plt.show()
```

### ***Answer 4:***
- ***Customer growth shows a `consistent and upward trajectory` over the entire period, `indicating strong and sustained` user acquisition without major drop-offs or volatility.***
- ***The growth pattern appears `nearly linear with slight acceleration` in later months, suggesting improved acquisition efficiency, possibly driven by optimized marketing strategies or increased brand awareness.***
- ***No significant seasonal dips or stagnation phases are observed, which reflects `stable demand and continuous engagement` across time.***

#  ***Business Recommendations***

Based on the analysis across revenue, customer behavior, delivery performance, product trends, and marketing effectiveness, the following strategic recommendations are proposed:

---

### **1. Optimize Delivery Operations**

- Delivery delays have a direct negative impact on customer ratings, indicating that timely delivery is critical for customer satisfaction.  
- The business should focus on improving last-mile delivery efficiency by optimizing delivery routes, enhancing partner performance monitoring, and reducing extreme delays.  
- Implementing real-time tracking and proactive delay communication can further improve customer experience.

---

### **2. Strengthen Customer Retention Strategy**

- A significant portion of revenue is driven by repeat and high-value (premium) customers. Retaining these customers should be a top priority.  
- Introduce loyalty programs, personalized offers, and targeted engagement strategies to increase customer lifetime value.  
- Re-engagement campaigns should be designed for inactive customers to reduce churn and improve retention rates.

---

### **3. Focus on High-Performing Products**

- A small group of products contributes disproportionately to total revenue, indicating strong product concentration.  
- Ensure consistent availability and visibility of these top-performing products through better inventory planning and promotions.  
- Additionally, expand the performance of underperforming products through bundling, pricing strategies, and cross-selling techniques.

---

### **4. Improve Marketing Effectiveness**

- Marketing campaigns show relatively similar ROAS, indicating stable but undifferentiated performance.  
- There is an opportunity to improve marketing efficiency by experimenting with campaign creatives, audience targeting, and channel strategies.  
- Budget allocation should be optimized by scaling consistently performing campaigns while refining or eliminating underperforming ones.

---

### **5. Increase Average Order Value**

- Customers tend to purchase a limited number of items per order, indicating potential to increase basket size.  
- Introduce strategies such as product bundling, volume discounts, and personalized recommendations to encourage higher spending per order.

---

### **6. Manage Demand Fluctuations**

- Order and revenue trends show variability over time, with weaker performance observed in certain periods.  
- To stabilize demand, the business should implement targeted promotions, seasonal campaigns, and strategic pricing during low-demand periods.

---

### **7. Adopt Data-Driven Decision Making**

- The analysis highlights the importance of integrating insights across multiple business areas including operations, customers, and marketing.  
- Continuous monitoring of KPIs such as delivery time, customer retention, and campaign performance can help in making informed and timely business decisions.
